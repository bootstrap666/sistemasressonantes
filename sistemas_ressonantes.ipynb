```python
import numpy as np
import cmath
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,100,0.01)

x = t * np.power(np.e, (-0.01+1j*2*np.pi)*t)
plt.plot(t,x.real)
```

```python
t = np.arange(0,1000,0.01)

x = t * np.power(np.e, (-0.01+1j*2*np.pi)*t)
plt.plot(t,x.real)
```
